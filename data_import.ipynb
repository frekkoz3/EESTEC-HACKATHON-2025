```python
import pandas as pd
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
```

```python
mat_data = loadmat('data 1\\degradation_data.mat')

data_dict = {}

for key in mat_data:
    if not key.startswith('__'):
        value = mat_data[key]
        if isinstance(value, (list, tuple)):
            try:
                arr = np.array(value)
                if arr.ndim == 1:
                    data_dict[key] = arr
            except:
                continue
        if isinstance(value, (int, float, complex)):
            data_dict[key] = [value]
        elif isinstance(value, np.ndarray):
            # Se è un vettore 1D, lo includiamo direttamente
            if value.ndim == 1:
                data_dict[key] = value
            # Se è 2D con una sola colonna o riga, lo convertiamo
            elif value.ndim == 2:
                if value.shape[0] == 1 or value.shape[1] == 1:
                    data_dict[key] = value.flatten()
            # Altrimenti lo saltiamo per evitare ambiguità
            else:
                pass

# Creiamo il DataFrame. Useremo la lunghezza minima comune per evitare errori di forma.
min_len = min(len(col) for col in data_dict.values())
trimmed_data = {k: v[:min_len] for k, v in data_dict.items()}
df = pd.DataFrame(trimmed_data)

df.head()
```

```python
df.to_csv('data 1\\degradation.csv', index=False)
```
